```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

# Parametr sigma
sigma = 1.0

# Symulacja szumu o rozkadzie Rayleigha
np.random.seed(42)  # Ustalenie losowego ziarna dla powtarzalnoci wynikw
samples_1000 = rayleigh.rvs(scale=sigma, size=1000)
samples_5000 = rayleigh.rvs(scale=sigma, size=5000)
samples_10000 = rayleigh.rvs(scale=sigma, size=10000)

# Funkcja pomocnicza do wizualizacji
def plot_histogram(samples, title):
    plt.figure(figsize=(8, 6))
    plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label="Symulacja")
    
    # Teoretyczna funkcja gstoci rozkadu Rayleigha
    x = np.linspace(0, np.max(samples), 1000)
    plt.plot(x, rayleigh.pdf(x, scale=sigma), 'r-', lw=2, label='Teoretyczna gsto')
    
    plt.title(title)
    plt.xlabel('Warto sigma')
    plt.ylabel('Gsto')
    plt.legend()
    plt.show()

# Wykresy dla rnych liczb prbek
plot_histogram(samples_1000, 'Rozkad Rayleigha - 1000 prbek')
plot_histogram(samples_5000, 'Rozkad Rayleigha - 5000 prbek')
plot_histogram(samples_10000, 'Rozkad Rayleigha - 10000 prbek')
```
