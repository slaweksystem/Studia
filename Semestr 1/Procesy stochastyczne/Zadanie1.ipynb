```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

# Parametr sigma
sigma = 1.0

# Symulacja szumu o rozkładzie Rayleigha
np.random.seed(42)  # Ustalenie losowego ziarna dla powtarzalności wyników
samples_1000 = rayleigh.rvs(scale=sigma, size=1000)
samples_5000 = rayleigh.rvs(scale=sigma, size=5000)
samples_10000 = rayleigh.rvs(scale=sigma, size=10000)

# Funkcja pomocnicza do wizualizacji
def plot_histogram(samples, title):
    plt.figure(figsize=(8, 6))
    plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label="Symulacja")
    
    # Teoretyczna funkcja gęstości rozkładu Rayleigha
    x = np.linspace(0, np.max(samples), 1000)
    plt.plot(x, rayleigh.pdf(x, scale=sigma), 'r-', lw=2, label='Teoretyczna gęstość')
    
    plt.title(title)
    plt.xlabel('Wartość sigma')
    plt.ylabel('Gęstość')
    plt.legend()
    plt.show()

# Wykresy dla różnych liczb próbek
plot_histogram(samples_1000, 'Rozkład Rayleigha - 1000 próbek')
plot_histogram(samples_5000, 'Rozkład Rayleigha - 5000 próbek')
plot_histogram(samples_10000, 'Rozkład Rayleigha - 10000 próbek')
```
